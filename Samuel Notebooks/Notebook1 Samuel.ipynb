```python
import re

# type 1: L/100 km or liters per 100 km
# type 2: km/L or km per L
# type 3: mpg lmp
# type 4: not available 
# type 5: others -> eliminate the number digits

#
def recognize_format_fuelX(input_data):
    types_counter = []
    
    # Conversion factors
    mpg_to_l_per_100km = 235.214583
    l_per_10km_to_l_per_100km = 10
    
    for val in input_data:
        if isinstance(val, str):  # Check if the value is a string
            if re.search(r'(L/100km|liters per 100 km)$', val):
                value = float(re.search(r'\d+\.\d+', val).group())
                print(value)
                if value != 0:  # Avoid division by zero
                    types_counter.append(100 / value)
                else:
                    types_counter.append(np.nan)
            
            elif re.search(r'(km/L|\bkm per L)$', val):
                value = float(re.search(r'\d+\.\d+', val).group())
                types_counter.append(value * 100)
            
            elif re.search(r'(\.mpg lmp\.|\bMPG \(AS\))$', val):
                value = float(re.search(r'\d+\.\d+', val).group())
                if value != 0:  # Avoid division by zero
                    types_counter.append(mpg_to_l_per_100km / value)
                else:
                    types_counter.append(np.nan)
            
            elif re.search(r'(\d+(\.\d+)?\s*L/10\s*km)$', val):
                value = float(re.search(r'\d+(\.\d+)?', val).group())
                if value != 0:  # Avoid division by zero
                    types_counter.append(1000 / (value / l_per_10km_to_l_per_100km))
                else:
                    types_counter.append(np.nan)
            elif val == 'zero':
                types_counter.append(0)
            
            else:
                types_counter.append(np.nan)  # For non-convertible values or NaN
        
        else:
            types_counter.append(np.nan)  # For non-string values
    
    return types_counter

# Example input data (replace this with your actual data)
fuel_consumption_column = fuelX['Fuel Consumption City']
print(fuel_consumption_column)
# Convert the input data
converted_data = recognize_format_fuelX(fuel_consumption_column )

# Create a DataFrame with the converted data
df_temp = pd.DataFrame({'Fuel_Consumption_Rate': converted_data})

# print(recognize_format_fuelX('32.72 MPG (AS)', counter))


# fuelX['Fuel Consumption City'].apply(lambda x : recognize_format_fuelX(x))
# fuelX['Fuel Consumption Hwy'].apply(lambda x : recognize_format_fuelX(x))
# fuelX['Fuel Consumption Comb'].apply(lambda x: recognize_format_fuelX(x))


        
```
